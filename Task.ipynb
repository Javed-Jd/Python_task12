```python
import numpy as np
```

```python

shapes = {
    1: (6,),
    2: (2, 3),
    3: (2, 3, 2),
    4: (2, 2, 2, 3),
    5: (2, 2, 2, 2, 2),
}

for dim, shape in shapes.items():
    arr = np.arange(np.prod(shape)).reshape(shape)

    # 5 attribute examples
    print("Attributes:")
    print("  shape     ", arr.shape)
    print("  ndim      ", arr.ndim)
    print("  size      ", arr.size)
    print("  dtype     ", arr.dtype)
    print("  itemsize  ", arr.itemsize)

    # 5 method examples
    print("Methods / examples:")
    print("  reshape(-1)       ", arr.reshape(-1))
    print("  transpose          ", np.transpose(arr))
    print("  sum (all)          ", arr.sum())
    print("  sum (axis=0)       ", arr.sum(axis=0))
    if arr.ndim >= 2:
        print("  mean (axis=1)      ", arr.mean(axis=1))
    else:
        print("  mean               ", arr.mean())
    print("  astype(float)      ", arr.astype(float))
```
